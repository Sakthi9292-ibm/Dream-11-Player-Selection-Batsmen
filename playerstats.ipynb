```python
def find_players(df,ls,city,team1,team2):
    a=df[df["City_Name"]==city]
    
    di={}
    for i in ls:
        
        b=a[a["Player_Id"]==i]
        
        #print(b)
        num=len(b["Batsman_Scored"])
        
        Runs=(b["Batsman_Scored"].sum())
       
                
        mn=b["Batsman_Scored"].mean()
        
        mdn=b["Batsman_Scored"].median()
        
        range_=(b["Batsman_Scored"].max())-(b["Batsman_Scored"].min())
        
        stde=b["Batsman_Scored"].std()
       
        if(mn!=0):
            cofv=stde/mn
        else:
            cofv=np.nan
        
        di[i]={"Player_Id":i,"Number of Matches":num,"Total Runs":Runs,"Mean":mn,"Median":mdn,"Range":range_,"Standard Deviation":stde,"Co. efficent of variance":cofv}
    #return di
    final=pa.DataFrame(di).T
    ff=final.merge(pda,on="Player_Id",how="inner").drop_duplicates()
    ff= ff[(ff["Team_Short_Code"]==team1) |(ff["Team_Short_Code"]==team2)]
    return ff
    
    
        
```

```python
ls = list(player_stats["Player_Id"].unique())
print(player_stats["City_Name"].unique())
print(pda["Team_Short_Code"].unique())
```

```python

play=find_players(player_stats,ls,"Bangalore","RCB","CSK")
```

```python

p=play[play["Mean"]>25].drop_duplicates()
print(p.shape)
p=p[["Player_Name","Team_Short_Code","Number of Matches","Total Runs","Mean","Median","Range","Standard Deviation","Co. efficent of variance"]]
p=p.set_index("Player_Name")
p.sort_values("Co. efficent of variance",ascending=True)
```
